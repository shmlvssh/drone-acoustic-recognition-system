## Р—Р°РҙР°РҪРёРө РәРҫРҪСҒСӮР°РҪСӮ

```python
# РһРұСҠРөРҙРёРҪСҸРөРј РІСҒРө РҝР°СҖР°РјРөСӮСҖСӢ РІ СҒСӮСҖСғРәСӮСғСҖСғ Config
class Config:
    # РҹР°СҖР°РјРөСӮСҖСӢ СҒРөСҒСҒРёРё
    SEED = 52                           # РЎРёРҙ РҙР»СҸ РІСҒРөС… РҝСҒРөРІРҙРҫСҒР»СғСҮР°Р№РҪСӢС… РҝРҫСҒР»РөРҙРҫРІР°СӮРөР»СҢРҪРҫСҒСӮРөР№

    # РҹР°СҖР°РјРөСӮСҖСӢ РҙР°СӮР°СҒРөСӮР°
    TRAIN_SIZE = 0.85                   # Р”РҫР»СҸ РҙР°РҪРҪСӢС… РҙР»СҸ РҫРұСғСҮРөРҪРёСҸ
    VAL_SIZE = 0.15                     # Р”РҫР»СҸ РҙР°РҪРҪСӢС… РҙР»СҸ РІР°Р»РёРҙР°СҶРёРё
    TEST_SIZE = 3000                    # Р”РҫР»СҸ РҙР°РҪРҪСӢС… РҙР»СҸ СӮРөСҒСӮРёСҖРҫРІР°РҪРёСҸ

    # РҹР°СҖР°РјРөСӮСҖСӢ Р°СғРҙРёРҫ
    SR = 32000                          # Р§Р°СҒСӮРҫСӮР° РҙРёСҒРәСҖРөСӮРёР·Р°СҶРёРё
    CHUNK_SEC = 1.0                     # Р”Р»РёРҪР° РҫРәРҪР° Р°РҪР°Р»РёР·Р°
    CHUNK_SAMPLES = int(SR * CHUNK_SEC) # РҡРҫР»РёСҮРөСҒСӮРІРҫ РҫСӮСҒСҮС‘СӮРҫРІ РІ РҫРәРҪРө

    # РҹР°СҖР°РјРөСӮСҖСӢ Р°СғРіРјРөРҪСӮР°СҶРёРё
    SNR_MIN = -5.0                      # РңРёРҪРёРјР°Р»СҢРҪСӢР№ SNR РҝСҖРё СҒРјРөСҲРёРІР°РҪРёРё РҙСҖРҫРҪР° СҒ СҲСғРјРҫРј
    SNR_MAX = 15.0                      # РңР°РәСҒРёРјР°Р»СҢРҪСӢР№ SNR
    GAIN_DB_MIN = -10.0                 # РқРёР¶РҪРёР№ РәСҖР°Р№ СҒР»СғСҮР°Р№РҪРҫРіРҫ СғСҒРёР»РөРҪРёСҸ (СғСҮРёСӮ РёРҪРІР°СҖРёР°РҪСӮРҪРҫСҒСӮРё Рә РіСҖРҫРјРәРҫСҒСӮРё)
    GAIN_DB_MAX = 10.0                  # Р’РөСҖС…РҪРёР№ РәСҖР°Р№

    # РӨРёР»СҢСӮСҖР°СҶРёСҸ СҮР°РҪРәРҫРІ РҝРҫ RMS (РҝСҖРёРјРөРҪСҸРөСӮСҒСҸ СӮРҫР»СҢРәРҫ Рә drone-С„Р°Р№Р»Р°Рј, label==1).
    # РЈРұРёСҖР°РөСӮ СғСӮРөСҮРәСғ РјРөСӮРәРё РҪР° СӮРёС…РёС…/РҝР°СғР·РҪСӢС… СғСҮР°СҒСӮРәР°С… Р·Р°РҝРёСҒРё.
    MIN_CHUNK_RMS = 1e-3

    # РЎРёРҪСӮРөСӮРёСҮРөСҒРәР°СҸ СӮРёСҲРёРҪР° вҖ” СҸРІРҪСӢР№ РҪРөРіР°СӮРёРІРҪСӢР№ РәР»Р°СҒСҒ
    SILENCE_DIR = "/tmp/dads/silence"
    NUM_SILENCE_FILES = 500             # РЎРәРҫР»СҢРәРҫ СҲСӮСғРә РіРөРҪРөСҖРёРј
    SILENCE_DURATION_SEC = 2.0          # Р”Р»РёСӮРөР»СҢРҪРҫСҒСӮСҢ РәР°Р¶РҙРҫРіРҫ С„Р°Р№Р»Р°

    # РҹР°СҖР°РјРөСӮСҖСӢ РҝСҖРөРҙРҫРұСҖР°РұРҫСӮРәРё РҙР°РҪРҪСӢС…
    FRAME_LENGTH = 1024                 # Р”Р»РёРҪР° РҫРәРҪР°
    FRAME_STEP = 512                    # РЁР°Рі РҫРәРҪР° (overlap 50%)
    NUM_FRAMES = 61
    FFT_SIZE = 1024                     # Р Р°Р·РјРөСҖ РҫРәРҪР° РҝСҖРөРҫРұСҖР°Р·РҫРІР°РҪРёСҸ РӨСғСҖСҢРө
    FMIN = 0.0                          # РқРёР¶РҪСҸСҸ РіСҖР°РҪРёСҶР° СҮР°СҒСӮРҫСӮ
    FMAX = 16000.0                      # Р’РөСҖС…РҪСҸСҸ РіСҖР°РҪРёСҶР° СҮР°СҒСӮРҫСӮ
    N_MELS = 96                         # РҡРҫР»РёСҮРөСҒСӮРІРҫ РјРөР»-С„РёР»СҢСӮСҖРҫРІ (РҝРҫРІСӢСҲРөРҪРҫ РҙР»СҸ Р»СғСҮСҲРөРіРҫ СҖР°Р·СҖРөСҲРөРҪРёСҸ РҪРёР¶РҪРёС… РіР°СҖРјРҫРҪРёРә РҙСҖРҫРҪР°)
    N_MFCC = 20                         # РҡРҫР»РёСҮРөСҒСӮРІРҫ РәРөРҝСҒСӮСҖР°Р»СҢРҪСӢС… РәРҫСҚС„С„РёСҶРёРөРҪСӮРҫРІ (РІСӢРұСҖР°РҪРҫ РәР°Рә РәРҫРјРҝСҖРҫРјРёСҒСҒ: Р»РҫРІРёСӮ РіР°СҖРјРҫРҪРёРәРё РҙСҖРҫРҪРҫРІ, РҪРө СҖР°Р·РҙСғРІР°СҸ 1D-CNN)

    # РҹР°СҖР°РјРөСӮСҖСӢ РҫРұСғСҮРөРҪРёСҸ
    BATCH_SIZE = 32                     # Р Р°Р·РјРөСҖ РҫРҙРҪРҫР№ РҝР°СҖСӮРёРё РҙР°РҪРҪСӢС… РҙР»СҸ РҫРұСғСҮРөРҪРёСҸ
    EPOCHS = 50                         # РҡРҫР»РёСҮРөСҒСӮРІРҫ СҚРҝРҫС… РҫРұСғСҮРөРҪРёСҸ
    STEPS_PER_EPOCH = 500               # РҡРҫР»РёСҮРөСҒСӮРІРҫ РұР°СӮСҮРөР№ РҪР° РҫРҙРҪСғ СҚРҝРҫС…Сғ (СғРІРөР»РёСҮРөРҪРҫ РҝРҫРҙ РҙРҫРҝРҫР»РҪРёСӮРөР»СҢРҪСӢРө Р°СғРіРјРөРҪСӮР°СҶРёРё)
    VAL_STEPS = 50                      # РҡРҫР»РёСҮРөСҒСӮРІРҫ РұР°СӮСҮРөР№ РҙР»СҸ РІР°Р»РёРҙР°СҶРёРё
```

## РЎРәР°СҮРёРІР°РҪРёРө РІРҪРөСҲРҪРөРіРҫ РҙР°СӮР°СҒРөСӮР° Drone Audio Detection Samples

## РЎРҫС…СҖР°РҪРөРҪРёРө СӮРөСҒСӮРҫРІРҫР№ РІСӢРұРҫСҖРәРё Drone Audio Detection Samples

```python
# РӨСғРҪРәСҶРёСҸ РҙР»СҸ СҒРҫС…СҖР°РҪРөРҪРёСҸ РҙР°СӮР°СҒРөСӮР° РІ С„РҫСҖРјР°СӮРө WAV-С„Р°Р№Р»РҫРІ
def export_dataset_to_wav(dataset, output_dir):
    for i, item in enumerate(tqdm(dataset, desc="РЎРҫС…СҖР°РҪРөРҪРёРө СӮРөСҒСӮРҫРІРҫР№ РІСӢРұРҫСҖРәРё")):
        # РҹРҫР»СғСҮР°РөРј СҒРІРҫР№СҒСӮРІР° Р°СғРҙРёРҫС„Р°Р№Р»Р° РёР· СҒР»РҫРІР°СҖСҸ
        audio_array = item['audio']['array']
        orig_sr = item['audio']['sampling_rate']
        label = item['label']

        # Р РөСҒРөРјРҝР»РёРј РҙРҫ СҶРөР»РөРІРҫР№ СҮР°СҒСӮРҫСӮСӢ
        if orig_sr != Config.SR:
            audio_array = librosa.resample(y=audio_array, orig_sr=orig_sr, target_sr=Config.SR)

        # РһРҝСҖРөРҙРөР»СҸРөРј РҝР°РҝРәСғ РҪР°Р·РҪР°СҮРөРҪРёСҸ
        folder = "drone" if label == 1 else "noise"
        filepath = f"{output_dir}/{folder}/sample_{i}.wav"

        # РЎРҫС…СҖР°РҪСҸРөРј РәР°Рә WAV-С„Р°Р№Р»
        sf.write(filepath, audio_array, Config.SR, subtype='PCM_16')

# РЎРҫР·РҙР°С‘Рј РҝР°РҝРәРё РҙР»СҸ СӮРөСҒСӮРҫРІСӢС… РҙР°РҪРҪСӢС…
test_dir = "/kaggle/working/test"
os.makedirs(f"{test_dir}/noise", exist_ok=True)
os.makedirs(f"{test_dir}/drone", exist_ok=True)

# РЎРҫС…СҖР°РҪСҸРөРј СӮРөСҒСӮРҫРІСӢРө РҙР°РҪРҪСӢРө РәР°Рә WAV-С„Р°Р№Р»СӢ РҙР»СҸ СҒРәР°СҮРёРІР°РҪРёСҸ
export_dataset_to_wav(TEST_DS, test_dir)

!zip -r -q test_dataset.zip /kaggle/working/test
```

```python
def dump_dataset_to_disk(hf_dataset, output_dir, desc_text):
    os.makedirs(f"{output_dir}/drone", exist_ok=True)
    os.makedirs(f"{output_dir}/noise", exist_ok=True)

    print(f"\nрҹ’ҫ РЎРҫС…СҖР°РҪСҸРөРј {desc_text} РІ {output_dir}...")

    # РҳСҒРҝРҫР»СҢР·СғРөРј РҫРұСӢСҮРҪСӢР№ СҶРёРәР». РқРёРәР°РәРҫРіРҫ multiprocessing, СҮСӮРҫРұСӢ РёР·РұРөР¶Р°СӮСҢ SegFault!
    for i, item in enumerate(tqdm(hf_dataset, desc=desc_text)):
        audio = item['audio']['array']
        label = item['label']
        folder = "drone" if label == 1 else "noise"
        filepath = f"{output_dir}/{folder}/{i}.wav"

        # soundfile.write СҖР°РұРҫСӮР°РөСӮ РҪР° C Рё СҒРҫС…СҖР°РҪСҸРөСӮ С„Р°Р№Р» Р·Р° РјРёРәСҖРҫСҒРөРәСғРҪРҙСӢ
        sf.write(filepath, audio, Config.SR, subtype='PCM_16')

TRAIN_DIR = "/tmp/dads/train"
VAL_DIR = "/tmp/dads/val"

dump_dataset_to_disk(TRAIN_DS, TRAIN_DIR, "РһРұСғСҮР°СҺСүСғСҺ РІСӢРұРҫСҖРәСғ (TRAIN)")
dump_dataset_to_disk(VAL_DS, VAL_DIR, "Р’Р°Р»РёРҙР°СҶРёРҫРҪРҪСғСҺ РІСӢРұРҫСҖРәСғ (VAL)")
```

## РӨСғРҪРәСҶРёРё РҙР»СҸ РҝСҖРөРҙРҫРұСҖР°РұРҫСӮРәРё РҙР°РҪРҪСӢС…

```python
# РҡРҫРҝРёСҸ РәР»Р°СҒСҒР° DataPreprocessor РёР· СӮРІРҫРөРіРҫ main.py
class DataPreprocessor:
    __slots__ = (
        "cfg", "_window", "_mel_weights", "_dct_matrix",
        "_mfcc_buf", "_mel_buf", "_rms_buf", "_frames_buf", "_spectrum_buf",
    )

    def __init__(self, cfg: Config):
        self.cfg = cfg
        self._window = np.hanning(cfg.FRAME_LENGTH).astype(np.float32)
        self._mel_weights = self._build_mel_filterbank(cfg)
        self._dct_matrix = self._build_dct_matrix(cfg)

        self._mfcc_buf = np.empty((1, cfg.NUM_FRAMES, cfg.N_MFCC), dtype=np.float32)
        self._mel_buf = np.empty((1, cfg.NUM_FRAMES, cfg.N_MELS, 1), dtype=np.float32)
        # log-RMS СҮР°РҪРәР° РәР°Рә РҙРҫРҝ.РІС…РҫРҙ 1D-CNN: РІРҫР·РІСҖР°СүР°РөСӮ СҒРөСӮРё Р°РұСҒРҫР»СҺСӮРҪСӢР№ СғСҖРҫРІРөРҪСҢ
        # СҒРёРіРҪР°Р»Р°, РәРҫСӮРҫСҖСӢР№ MFCC РҫСӮРұСҖР°СҒСӢРІР°СҺСӮ СҮРөСҖРөР· DCT (Р° random_gain РҙРөР»Р°РөСӮ
        # MFCC[0] РҪРөРёРҪС„РҫСҖРјР°СӮРёРІРҪСӢРј).
        self._rms_buf = np.empty((1, 1), dtype=np.float32)
        self._frames_buf = np.empty((cfg.NUM_FRAMES, cfg.FRAME_LENGTH), dtype=np.float32)
        self._spectrum_buf = np.empty((cfg.NUM_FRAMES, cfg.FFT_SIZE // 2 + 1), dtype=np.float32)

    @staticmethod
    def _build_mel_filterbank(cfg: Config) -> np.ndarray:
        hz_to_mel = lambda hz: 2595.0 * np.log10(1.0 + hz / 700.0)
        mel_to_hz = lambda mel: 700.0 * (10.0 ** (mel / 2595.0) - 1.0)
        num_bins = cfg.FFT_SIZE // 2 + 1
        mel_points = np.linspace(hz_to_mel(cfg.FMIN), hz_to_mel(cfg.FMAX), cfg.N_MELS + 2)
        hz_points = mel_to_hz(mel_points)
        bins = np.floor((cfg.FFT_SIZE + 1) * hz_points / cfg.SR).astype(int)

        filters = np.zeros((num_bins, cfg.N_MELS), dtype=np.float32)
        for i in range(cfg.N_MELS):
            left, center, right = bins[i], bins[i + 1], bins[i + 2]
            if center > left: filters[left:center, i] = np.arange(center - left) / (center - left)
            if right > center: filters[center:right, i] = np.arange(right - center, 0, -1) / (right - center)
        return filters

    @staticmethod
    def _build_dct_matrix(cfg: Config) -> np.ndarray:
        n = np.arange(cfg.N_MELS)
        k = np.arange(cfg.N_MFCC)
        dct = np.cos(np.pi * k[np.newaxis, :] * (2 * n[:, np.newaxis] + 1) / (2 * cfg.N_MELS))
        dct *= np.sqrt(2.0 / cfg.N_MELS)
        return dct.astype(np.float32)

    def _frame(self, audio: np.ndarray) -> np.ndarray:
        audio = np.ascontiguousarray(audio)
        byte_step = audio.strides[0]
        return as_strided(audio, shape=(self.cfg.NUM_FRAMES, self.cfg.FRAME_LENGTH),
                          strides=(self.cfg.FRAME_STEP * byte_step, byte_step))

    def preprocess(self, audio: np.ndarray):
        if audio.dtype != np.float32: audio = audio.astype(np.float32, copy=False)
        # log-RMS СҒСҮРёСӮР°РөРј Р”Рһ РјР°СӮСҖРёСҮРҪСӢС… РҫРҝРөСҖР°СҶРёР№. log compress'РёСӮ РұРҫР»СҢСҲРҫР№
        # РҙРёРҪР°РјРёСҮРөСҒРәРёР№ РҙРёР°РҝР°Р·РҫРҪ (1e-5..1.0) РІ Р»РёРҪРөР№РҪСӢР№ (-5..0), СғРҙРҫРұРҪСӢР№ РҙР»СҸ Dense.
        rms_raw = float(np.sqrt(np.mean(audio * audio)))
        self._rms_buf[0, 0] = np.log(rms_raw + 1e-7)
        np.multiply(self._frame(audio), self._window, out=self._frames_buf)
        rfft_out = np.fft.rfft(self._frames_buf, n=self.cfg.FFT_SIZE)
        np.multiply(rfft_out.real, rfft_out.real, out=self._spectrum_buf)
        self._spectrum_buf += rfft_out.imag * rfft_out.imag
        np.sqrt(self._spectrum_buf, out=self._spectrum_buf)
        mel_view = self._mel_buf[0, :, :, 0]
        np.matmul(self._spectrum_buf, self._mel_weights, out=mel_view)
        mel_view += 1e-6
        np.log(mel_view, out=mel_view)
        np.matmul(mel_view, self._dct_matrix, out=self._mfcc_buf[0])
        return self._mfcc_buf, self._mel_buf, self._rms_buf


# РҘСҖР°РҪРёР»РёСүРө РҙР»СҸ РҝРҫСӮРҫРәРҫРұРөР·РҫРҝР°СҒРҪРҫРіРҫ РёРҪСҒСӮР°РҪСҶРёСҖРҫРІР°РҪРёСҸ РҝСҖРөРҝСҖРҫСҶРөСҒСҒРҫСҖР°
thread_local_storage = threading.local()

def get_thread_local_preprocessor():
    if not hasattr(thread_local_storage, "preproc"):
        thread_local_storage.preproc = DataPreprocessor(Config)
    return thread_local_storage.preproc

def numpy_preprocess_wrapper(audio_np):
    preproc = get_thread_local_preprocessor()
    mfcc, mel, rms = preproc.preprocess(audio_np)
    # РһРұСҸР·Р°СӮРөР»СҢРҪРҫ РІРҫР·РІСҖР°СүР°РөРј РәРҫРҝРёРё, СӮР°Рә РәР°Рә preproc РҝРөСҖРөРёСҒРҝРҫР»СҢР·СғРөСӮ РҝР°РјСҸСӮСҢ!
    # Рҳ РұРөСҖРөРј [0], СҮСӮРҫРұСӢ СғРұСҖР°СӮСҢ СҖР°Р·РјРөСҖРҪРҫСҒСӮСҢ РұР°СӮСҮР° (РәРҫСӮРҫСҖСғСҺ РҙРҫРұР°РІР»СҸР» DataPreprocessor).
    return mfcc[0].copy(), mel[0].copy(), rms[0].copy()

@tf.function
def mix_noise(signal, noise, target_snr):
    # РЎСҮРёСӮР°РөРј RMS РҝРҫР»РөР·РҪРҫРіРҫ СҒРёРіРҪР°Р»Р° Рё СҲСғРјР°
    rms_signal = tf.sqrt(tf.reduce_mean(tf.square(signal)))
    rms_noise = tf.sqrt(tf.reduce_mean(tf.square(noise)))

    # РҳР·РұРөРіР°РөРј РҙРөР»РөРҪРёСҸ РҪР° РҪРҫР»СҢ
    rms_noise = tf.maximum(rms_noise, 1e-7)

    # РЎСҮРёСӮР°РөРј RMS СҲСғРјР° РҙР»СҸ СҶРөР»РөРІРҫРіРҫ SNR
    snr_factor = tf.pow(10.0, target_snr / 20.0)
    target_rms_noise = rms_signal / snr_factor

    # РңР°СҒСҲСӮР°РұРёСҖСғРөРј СҲСғРј
    noise_scaled = noise * (target_rms_noise / rms_noise)

    # РЎРјРөСҲРёРІР°РөРј
    mixed = signal + noise_scaled

    # РңРёРҪРёРјР°РәСҒРҪР°СҸ РҪРҫСҖРјР°Р»РёР·Р°СҶРёСҸ
    max_val = tf.reduce_max(tf.abs(mixed))
    mixed_normalized = tf.cond(
        max_val > 0,
        lambda: mixed / max_val,
        lambda: mixed
    )
    return mixed_normalized

# Р—Р°РјРөРҪСҸРөРј СҒСӮР°СҖСӢР№ extract_features РҪР° РІСӢР·РҫРІ РҪР°СҲРөР№ numpy-С„СғРҪРәСҶРёРё
@tf.function
def extract_features(audio_tensor):
    # tf.numpy_function РҝРҫР·РІРҫР»СҸРөСӮ РІСӢР·РІР°СӮСҢ Python/NumPy РәРҫРҙ РІРҪСғСӮСҖРё TF-РіСҖР°С„Р°
    mfcc, mel, rms = tf.numpy_function(
        numpy_preprocess_wrapper,
        [audio_tensor], [tf.float32, tf.float32, tf.float32]
    )
    # Р’РҫСҒСҒСӮР°РҪР°РІР»РёРІР°РөРј С„РҫСҖРјСғ СӮРөРҪР·РҫСҖРҫРІ (tf.numpy_function СҒСӮРёСҖР°РөСӮ shape)
    mfcc.set_shape((Config.NUM_FRAMES, Config.N_MFCC))
    mel.set_shape((Config.NUM_FRAMES, Config.N_MELS, 1))
    rms.set_shape((1,))

    return mfcc, mel, rms

# РһСҒСӮР°Р»СҢРҪРҫР№ РәРҫРҙ augment_and_extract РҫСҒСӮР°РөСӮСҒСҸ РұРөР· РёР·РјРөРҪРөРҪРёР№,
# СӮРҫР»СҢРәРҫ СғРұРёСҖР°РөРј СҖСғСҮРҪРҫРө РҙРҫРұР°РІР»РөРҪРёРө СҖР°Р·РјРөСҖРҪРҫСҒСӮРё РәР°РҪР°Р»Р° РҙР»СҸ mels,
# СӮР°Рә РәР°Рә РҪР°СҲ DataPreprocessor СғР¶Рө РҫСӮРҙР°С‘СӮ (frames, mels, 1)

# Speed perturbation = pitch + tempo РҫРҙРҪРҫРІСҖРөРјРөРҪРҪРҫ. РӣРёРҪРөР№РҪР°СҸ РёРҪСӮРөСҖРҝРҫР»СҸСҶРёСҸ СҮРөСҖРөР·
# tf.linspace + tf.gather вҖ” РұРөР· librosa. Р РөР°Р»РёСҒСӮРёСҮРҪРҫ РёРјРёСӮРёСҖСғРөСӮ Р”РҫРҝР»РөСҖ-СҒРҙРІРёРі
# РҙРІРёР¶СғСүРөРіРҫСҒСҸ РёСҒСӮРҫСҮРҪРёРәР° (BPF-РіР°СҖРјРҫРҪРёРәРё СҒСҠРөР·Р¶Р°СҺСӮ РҝРҫ СҮР°СҒСӮРҫСӮРө).
@tf.function
def pitch_shift(audio, factor):
    n = tf.shape(audio)[0]
    n_f = tf.cast(n, tf.float32)
    new_n = tf.cast(n_f / factor, tf.int32)
    indices = tf.linspace(0.0, n_f - 1.0, new_n)
    floor_idx = tf.cast(indices, tf.int32)
    ceil_idx = tf.minimum(floor_idx + 1, n - 1)
    frac = indices - tf.cast(floor_idx, tf.float32)
    shifted = tf.gather(audio, floor_idx) * (1.0 - frac) + tf.gather(audio, ceil_idx) * frac
    # Trim РёР»Рё pad РҙРҫ РёСҒС…РҫРҙРҪРҫР№ РҙР»РёРҪСӢ, СҮСӮРҫРұСӢ downstream РІРёРҙРөР» С„РёРәСҒРёСҖРҫРІР°РҪРҪСӢР№ shape.
    shifted = tf.cond(
        new_n >= n,
        lambda: shifted[:n],
        lambda: tf.concat([shifted, tf.zeros(n - new_n, dtype=tf.float32)], axis=0),
    )
    shifted.set_shape((Config.CHUNK_SAMPLES,))
    return shifted


# РЎР»СғСҮР°Р№РҪСӢР№ low-pass СҮРөСҖРөР· FFT СҒ РіР»Р°РҙРәРёРј roll-off. РҳРјРёСӮРёСҖСғРөСӮ Р°СӮРјРҫСҒС„РөСҖРҪРҫРө
# РҝРҫРіР»РҫСүРөРҪРёРө: РІСӢСҒРҫРәРёРө СҮР°СҒСӮРҫСӮСӢ Р·Р°СӮСғС…Р°СҺСӮ СҚРәСҒРҝРҫРҪРөРҪСҶРёР°Р»СҢРҪРҫ СҒ СҖР°СҒСҒСӮРҫСҸРҪРёРөРј. Cutoff
# РІ log-uniform РҙРёР°РҝР°Р·РҫРҪРө РҝРҫРәСҖСӢРІР°РөСӮ РҙР°Р»СҢРҪРҫСҒСӮРё РҫСӮ РұР»РёР·РәРҫР№ Р·Р°РҝРёСҒРё (16 РәР“СҶ)
# РҙРҫ РҙРөСҒСҸСӮРәРҫРІ РјРөСӮСҖРҫРІ (3 РәР“СҶ).
@tf.function
def random_lowpass(audio, min_cutoff_hz, max_cutoff_hz):
    n = Config.CHUNK_SAMPLES
    spec = tf.signal.rfft(audio)
    n_bins = tf.shape(spec)[0]
    freqs = tf.cast(tf.range(n_bins), tf.float32) * Config.SR / tf.cast(n, tf.float32)
    log_min = tf.math.log(tf.cast(min_cutoff_hz, tf.float32))
    log_max = tf.math.log(tf.cast(max_cutoff_hz, tf.float32))
    cutoff = tf.exp(tf.random.uniform([], log_min, log_max))
    # РҹР»Р°РІРҪСӢР№ roll-off РІ exp РІРјРөСҒСӮРҫ Р¶С‘СҒСӮРәРҫРіРҫ СҒСӮСғРҝРөРҪСҮР°СӮРҫРіРҫ СҒСҖРөР·Р° вҖ” РұРөР· ringing.
    rolloff_hz = 1000.0
    gain = tf.where(freqs < cutoff, 1.0, tf.exp(-(freqs - cutoff) / rolloff_hz))
    spec_filtered = spec * tf.cast(gain, tf.complex64)
    out = tf.signal.irfft(spec_filtered, fft_length=[n])
    out.set_shape((Config.CHUNK_SAMPLES,))
    return out


@tf.function
def augment_and_extract(target, noise_audio, is_training=True):
    target_audio, label = target

    if is_training:
        result_audio = tf.cond(
            tf.equal(label, 1.0),
            lambda: mix_noise(
                target_audio,
                noise_audio,
                target_snr=tf.random.uniform([], minval=Config.SNR_MIN, maxval=Config.SNR_MAX)
            ),
            lambda: target_audio
        )

        do_pitch = tf.random.uniform([]) < 0.5
        result_audio = tf.cond(
            do_pitch,
            lambda: pitch_shift(result_audio, tf.random.uniform([], 0.95, 1.05)),
            lambda: result_audio,
        )

        do_lp = tf.logical_and(tf.equal(label, 1.0), tf.random.uniform([]) < 0.5)
        result_audio = tf.cond(
            do_lp,
            lambda: random_lowpass(result_audio, 3000.0, 16000.0),
            lambda: result_audio,
        )
    else:
        result_audio = target_audio

    # === РҡР РҳРўРҳР§Р•РЎРҡРҳРҷ РӨРҳРҡРЎ: РЈРқРҳР’Р•Р РЎРҗРӣР¬РқРҗРҜ РқРһР РңРҗРӣРҳР—РҗРҰРҳРҜ ===
    # РҡР°Рә РІ РұРҫРөРІРҫРј main.py, РјСӢ РҫРұСҸР·Р°РҪСӢ РҪРҫСҖРјР°Р»РёР·РҫРІР°СӮСҢ Р’РЎР• СҮР°РҪРәРё!
    max_val = tf.reduce_max(tf.abs(result_audio))
    result_audio = tf.cond(
        max_val > 1e-5, # Р—Р°СүРёСӮР° РҫСӮ РҙРөР»РөРҪРёСҸ РҪР° СҮРёСҒСӮСӢР№ РҪРҫР»СҢ (СҒРёРҪСӮРөСӮРёСҮРөСҒРәСғСҺ СӮРёСҲРёРҪСғ)
        lambda: result_audio / max_val,
        lambda: result_audio
    )

    mfcc, mel_spectrogram, rms = extract_features(result_audio)
    return {"mfcc": mfcc, "mel_spectrogram": mel_spectrogram, "rms": rms}, label
    
```

## РӨСғРҪРәСҶРёРё РҙР»СҸ Р·Р°РіСҖСғР·РәРё РҙР°РҪРҪСӢС…

```python
def read_wav(file_path):
    path = file_path.decode('utf-8')
    try:
        # soundfile (libsndfile) "РІСҒРөСҸРҙРөРҪ" Рё РҝСҖРҫСҮРёСӮР°РөСӮ Format 3 РұРөР· РҝСҖРҫРұР»РөРј
        audio, sr = sf.read(path, dtype='float32')

        # Р•СҒР»Рё Р·РІСғРә СҒСӮРөСҖРөРҫ - СҒРІРҫРҙРёРј РІ РјРҫРҪРҫ
        if audio.ndim > 1:
            audio = np.mean(audio, axis=1)

        # РқР° РІСҒСҸРәРёР№ СҒР»СғСҮР°Р№: РөСҒР»Рё РІ DEMAND СҮР°СҒСӮРҫСӮР° РҪРө 32РәР“СҶ, СҖРөСҒРөРјРҝР»РёРј РҪР° Р»РөСӮСғ
        if sr != Config.SR:
            audio = librosa.resample(y=audio, orig_sr=sr, target_sr=Config.SR)

        return audio.astype(np.float32)
    except Exception as e:
        print(f"\nвҡ пёҸ РһСҲРёРұРәР° СҮСӮРөРҪРёСҸ С„Р°Р№Р»Р° {path}: {e}. Р’РҫР·РІСҖР°СүР°РөРј СӮРёСҲРёРҪСғ.")
        # Р§СӮРҫРұСӢ РҝР°Р№РҝР»Р°Р№РҪ РҪРө СғРҝР°Р» РёР·-Р·Р° РҫРҙРҪРҫРіРҫ РұРёСӮРҫРіРҫ С„Р°Р№Р»Р°, РІРҫР·РІСҖР°СүР°РөРј РәСғСҒРҫРә СӮРёСҲРёРҪСӢ
        return np.zeros(Config.CHUNK_SAMPLES, dtype=np.float32)

# Р—Р°СҶРёРәР»РёРІР°РҪРёРө РәРҫСҖРҫСӮРәРёС… С„Р°Р№Р»РҫРІ: РІРјРөСҒСӮРҫ РҫСӮРұСҖР°СҒСӢРІР°РҪРёСҸ (pad_end=False СӮРөСҖСҸР»
# РәРҫСҖРҫСӮРәРёРө drone-Р·Р°РҝРёСҒРё) РёР»Рё РҝР°РҙРҙРёРҪРіР° РҪСғР»СҸРјРё (Р·Р°СҲСғРјР»СҸРөСӮ СҖР°СҒРҝСҖРөРҙРөР»РөРҪРёРө вҖ”
# РјРҫРҙРөР»СҢ СғСҮРёР»Р°СҒСҢ РұСӢ В«СҖРөР°Р»СҢРҪСӢР№ СҒРёРіРҪР°Р» + СӮРёСҲРёРҪР° -> droneВ») РҝРҫРІСӮРҫСҖСҸРөРј СҒРҫРҙРөСҖР¶РёРјРҫРө
# np.tile-Р°РҪР°Р»РҫРіРҫРј РҙРҫ РҙР»РёРҪСӢ CHUNK_SAMPLES.
def _loop_short_audio(audio):
    n = tf.shape(audio)[0]
    target = tf.constant(Config.CHUNK_SAMPLES, dtype=tf.int32)
    # ceil(target / n)
    repeats = (target + n - 1) // n
    looped = tf.tile(audio, [repeats])
    return looped[:target]


# РӨСғРҪРәСҶРёСҸ РҙР»СҸ СҮСӮРөРҪРёСҸ WAV-С„Р°Р№Р»Р° Рё РҪР°СҖРөР·РәРё РөРіРҫ РҪР° СҮР°РҪРәРё
def load_and_chunk_wav(file_path, label):
    audio = tf.numpy_function(read_wav, [file_path], tf.float32)
    audio.set_shape([None])

    # Р•СҒР»Рё С„Р°Р№Р» РәРҫСҖРҫСҮРө РҫРәРҪР° вҖ” Р·Р°СҶРёРәР»РёРІР°РөРј РҙРҫ РҝРҫР»РҪРҫРіРҫ РҫРәРҪР°. РҳРҪР°СҮРө tf.signal.frame
    # СҒ pad_end=False РІРөСҖРҪСғР» РұСӢ 0 СҮР°РҪРәРҫРІ (СҒСӮР°СҖРҫРө РҝРҫРІРөРҙРөРҪРёРө), Рё РәРҫСҖРҫСӮРәРёРө
    # drone-Р·Р°РҝРёСҒРё СӮРөСҖСҸР»РёСҒСҢ РұСӢ РҝРҫР»РҪРҫСҒСӮСҢСҺ.
    audio = tf.cond(
        tf.shape(audio)[0] < Config.CHUNK_SAMPLES,
        lambda: _loop_short_audio(audio),
        lambda: audio,
    )

    # pad_end=False: РҪРө РҙРҫРҝРёСҒСӢРІР°РөРј РҪСғР»Рё РІ РәРҫРҪРөСҶ. РҳРҪР°СҮРө С…РІРҫСҒСӮРҫРІРҫР№ СҮР°РҪРә РҙСҖРҫРҪ-С„Р°Р№Р»Р°
    # РұСӢР» РұСӢ РҪР° 30-90% РёР· РҪСғР»РөР№ Рё РҝРҫРҝР°РҙР°Р» РІ РҫРұСғСҮРөРҪРёРө СҒ РјРөСӮРәРҫР№ 1 вҖ” СҒРөСӮСҢ СғСҮРёР»Р°СҒСҢ РұСӢ
    # В«zeros -> droneВ». РӨР°Р№Р»СӢ РҙР»РёРҪРҫР№ РәСҖР°СӮРҪРҫР№ CHUNK_SAMPLES // 2 РҝСҖРё СҚСӮРҫРј
    # РҫРұСҖР°РұР°СӮСӢРІР°СҺСӮСҒСҸ РҝРҫР»РҪРҫСҒСӮСҢСҺ РұР»Р°РіРҫРҙР°СҖСҸ Р·Р°СҶРёРәР»РёРІР°РҪРёСҺ РІСӢСҲРө.
    # frame_step = CHUNK_SAMPLES // 2 РҙР°С‘СӮ 50% overlap: С„Р°РәСӮРёСҮРөСҒРәРёР№ РҫРұСҠС‘Рј
    # СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪСӢС… СҒСҚРјРҝР»РҫРІ СғРҙРІР°РёРІР°РөСӮСҒСҸ РұРөР· РҪРҫРІСӢС… Р·Р°РҝРёСҒРөР№. Р”СҖРҫРҪ РёР»Рё РәРҫСҖРҫСӮРәРҫРө
    # СҲСғРјРҫРІРҫРө СҒРҫРұСӢСӮРёРө РјРҫР¶РөСӮ РҫРәР°Р·Р°СӮСҢСҒСҸ РІ Р»СҺРұРҫР№ СҮР°СҒСӮРё РҫРәРҪР°, СҮСӮРҫ СғР»СғСҮСҲР°РөСӮ
    # РҝРҫР·РёСҶРёРҫРҪРҪСғСҺ РёРҪРІР°СҖРёР°РҪСӮРҪРҫСҒСӮСҢ.
    chunks = tf.signal.frame(
        audio,
        frame_length=Config.CHUNK_SAMPLES,
        frame_step=Config.CHUNK_SAMPLES // 2,
        pad_end=False
    )

    # Р”Р»СҸ drone-С„Р°Р№Р»РҫРІ (label==1) РҙРҫРҝРҫР»РҪРёСӮРөР»СҢРҪРҫ РҫСӮРұСҖР°СҒСӢРІР°РөРј СӮРёС…РёРө СҮР°РҪРәРё вҖ”
    # СҚСӮРҫ РҝР°СғР·СӢ РІРҪСғСӮСҖРё Р·Р°РҝРёСҒРё, РјРҫРјРөРҪСӮ РҙРҫ РІР·Р»С‘СӮР° Рё СӮ.Рҝ. РҳС… РҪРөР»СҢР·СҸ РҝРҫРјРөСҮР°СӮСҢ РәР°Рә
    # В«РҙСҖРҫРҪВ», РёРҪР°СҮРө СғСӮРөСҮРәР° РјРөСӮРәРё. РЁСғРјРҫРІСӢРө С„Р°Р№Р»СӢ РҝСҖРҫРҝСғСҒРәР°РөРј РұРөР· С„РёР»СҢСӮСҖР°:
    # СӮРёС…РёРө СҒСҚРјРҝР»СӢ СӮР°Рј РәРҫСҖСҖРөРәСӮРҪРҫ РҪРөСҒСғСӮ РјРөСӮРәСғ 0 Рё РҪСғР¶РҪСӢ РәР°Рә СҸРІРҪСӢР№ РҪРөРіР°СӮРёРІРҪСӢР№ РәР»Р°СҒСҒ
    # (СҒРёРҪСӮРөСӮРёСҮРөСҒРәР°СҸ СӮРёСҲРёРҪР° РёР· cell 17 РёРҙС‘СӮ РҝРҫ СҚСӮРҫРјСғ Р¶Рө РҝСғСӮРё).
    def _filter_drone_chunks():
        rms = tf.sqrt(tf.reduce_mean(tf.square(chunks), axis=1))
        return tf.boolean_mask(chunks, rms > Config.MIN_CHUNK_RMS)

    chunks = tf.cond(
        tf.equal(label, 1.0),
        _filter_drone_chunks,
        lambda: chunks
    )

    num_chunks = tf.shape(chunks)[0]
    chunk_ds = tf.data.Dataset.from_tensor_slices(chunks)
    label_ds = tf.data.Dataset.from_tensors(label).repeat(tf.cast(num_chunks, dtype=tf.int64))
    return tf.data.Dataset.zip((chunk_ds, label_ds))

# РҹРҫСӮРҫРә СҮР°РҪРәРҫРІ С„РёРәСҒРёСҖРҫРІР°РҪРҪРҫР№ РјРөСӮРәРё. РӨР°Р№Р»СӢ РҝРөСҖРө-СҲР°С„Р»СҸСӮСҒСҸ РәР°Р¶РҙСғСҺ РёСӮРөСҖР°СҶРёСҺ,
# interleave round-robin'РёСӮ СҮР°РҪРәРё РёР· 16 С„Р°Р№Р»РҫРІ РҫРҙРҪРҫРІСҖРөРјРөРҪРҪРҫ. Р’РҪСғСӮСҖРё РҫРҙРҪРҫРіРҫ РҝРҫСӮРҫРәР°
# РұР°Р»Р°РҪСҒ РҝРҫ РәР»Р°СҒСҒР°Рј РҪРө РҪСғР¶РөРҪ вҖ” РІСҒРө СҮР°РҪРәРё РёРјРөСҺСӮ РҫРҙРҪСғ Рё СӮСғ Р¶Рө РјРөСӮРәСғ.
def _build_class_stream(file_paths, label_value):
    paths_ds = tf.data.Dataset.from_tensor_slices(list(file_paths))
    paths_ds = paths_ds.shuffle(
        buffer_size=len(file_paths),
        reshuffle_each_iteration=True,
    ).repeat()
    return paths_ds.interleave(
        lambda p: load_and_chunk_wav(p, label_value),
        cycle_length=16,
        num_parallel_calls=tf.data.AUTOTUNE,
    )


# РЎРұРҫСҖ С„РёРҪР°Р»СҢРҪРҫРіРҫ tf.data-РәРҫРҪРІРөР№РөСҖР°. Р“Р»Р°РІРҪСӢР№ С„РёРәСҒ РҫСӮРҪРҫСҒРёСӮРөР»СҢРҪРҫ РҝСҖРөР¶РҪРөР№ РІРөСҖСҒРёРё:
# СҖР°РҪСҢСҲРө interleave РҙР°РІР°Р» РҝРҫСӮРҫРә СҮР°РҪРәРҫРІ, Р’Р—Р’Р•РЁР•РқРқР«Рҷ РҝРҫ РҙР»РёРҪР°Рј С„Р°Р№Р»РҫРІ (РҙР»РёРҪРҪСӢРө
# DEMAND-С„Р°Р№Р»СӢ РҙРҫРјРёРҪРёСҖРҫРІР°Р»Рё ~7:1), Рё class_weight СӮРҫР»СҢРәРҫ СҮР°СҒСӮРёСҮРҪРҫ РәРҫРјРҝРөРҪСҒРёСҖРҫРІР°Р»
# СҚСӮРҫ РҪР° СғСҖРҫРІРҪРө Р»РҫСҒСҒР°. РўРөРҝРөСҖСҢ sample_from_datasets СҒ РІРөСҒР°РјРё 0.5/0.5 РіР°СҖР°РҪСӮРёСҖСғРөСӮ
# РұР°Р»Р°РҪСҒ РҪР° СҮР°РҪРә-СғСҖРҫРІРҪРө вҖ” class_weight РұРҫР»СҢСҲРө РҪРө РҪСғР¶РөРҪ.
def create_dataset_from_wav(drone_files, noise_files, is_training=True):
    if is_training:
        # Р”РІР° РҪРөР·Р°РІРёСҒРёРјСӢС… РҝРҫСӮРҫРәР°. РҡР°Р¶РҙСӢР№ РұРөСҒРәРҫРҪРөСҮРҪСӢР№ (repeat РІРҪСғСӮСҖРё _build_class_stream).
        drone_ds = _build_class_stream(drone_files, 1.0)
        noise_ds = _build_class_stream(noise_files, 0.0)

        # 50/50 РҪР° СҮР°РҪРә-СғСҖРҫРІРҪРө, РҪРөР·Р°РІРёСҒРёРјРҫ РҫСӮ РҙР»РёРҪ С„Р°Р№Р»РҫРІ.
        main_ds = tf.data.Dataset.sample_from_datasets(
            [drone_ds, noise_ds],
            weights=[0.5, 0.5],
            stop_on_empty_dataset=False,
        )
        # Р”РҫРҝРҫР»РҪРёСӮРөР»СҢРҪСӢР№ shuffle СғРұРёСҖР°РөСӮ Р»РҫРәР°Р»СҢРҪСғСҺ СҖРөРіСғР»СҸСҖРҪРҫСҒСӮСҢ СҮРөСҖРөРҙРҫРІР°РҪРёСҸ.
        main_ds = main_ds.shuffle(buffer_size=5000, reshuffle_each_iteration=True)

        # РҹР°СҖР°Р»Р»РөР»СҢРҪСӢР№ РҝРҫСӮРҫРә СҲСғРјР° РҙР»СҸ mix_noise (С„РҫРҪ РҝРҫРҙ РҙСҖРҫРҪ).
        bg_noise_ds = _build_class_stream(noise_files, 0.0).map(
            lambda audio, _label: audio
        )
        bg_noise_ds = bg_noise_ds.shuffle(buffer_size=1000)

        zipped_ds = tf.data.Dataset.zip((main_ds, bg_noise_ds))
    else:
        # Р’Р°Р»РёРҙР°СҶРёСҸ: РұР°Р»Р°РҪСҒ РҪРө РҪСғР¶РөРҪ вҖ” РјРөСӮСҖРёРәРё РҙРҫР»Р¶РҪСӢ РҫСӮСҖР°Р¶Р°СӮСҢ СҖРөР°Р»СҢРҪРҫРө
        # СҖР°СҒРҝСҖРөРҙРөР»РөРҪРёРө РІР°Р»РёРҙР°СҶРёРҫРҪРҪРҫР№ РІСӢРұРҫСҖРәРё. РҳРҙС‘Рј РҝСҖРөР¶РҪРёРј РҝСғСӮС‘Рј СҮРөСҖРөР· interleave.
        all_files = list(drone_files) + list(noise_files)
        all_labels = [1.0] * len(drone_files) + [0.0] * len(noise_files)
        rng = np.random.default_rng(Config.SEED)
        order = rng.permutation(len(all_files))
        all_files = [all_files[i] for i in order]
        all_labels = [all_labels[i] for i in order]

        source_ds = tf.data.Dataset.from_tensor_slices((all_files, all_labels))
        main_ds = source_ds.interleave(
            load_and_chunk_wav,
            cycle_length=16,
            num_parallel_calls=tf.data.AUTOTUNE,
        )
        dummy_noise = tf.data.Dataset.from_tensors(
            tf.zeros(Config.CHUNK_SAMPLES, dtype=tf.float32)
        ).repeat()
        zipped_ds = tf.data.Dataset.zip((main_ds, dummy_noise))

    final_ds = zipped_ds.map(
        lambda main_data, bg_noise: augment_and_extract(main_data, bg_noise, is_training),
        num_parallel_calls=tf.data.AUTOTUNE,
    )

    final_ds = final_ds.batch(Config.BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

    if is_training:
        final_ds = final_ds.repeat()

    return final_ds

# РӨСғРҪРәСҶРёСҸ РҙР»СҸ СҮСӮРөРҪРёСҸ РҙР°РҪРҪСӢС… РёР· HF-РҙР°СӮР°СҒРөСӮР°
def dataset_generator(dataset):
    for item in dataset:
        # РҹРҫР»СғСҮР°РөРј СҒРІРҫР№СҒСӮРІР° Р°СғРҙРёРҫС„Р°Р№Р»Р° РёР· СҒР»РҫРІР°СҖСҸ
        audio = item['audio']['array']
        orig_sr = item['audio']['sampling_rate']
        label = item['label']

        # Р РөСҒРөРјРҝР»РёРј РҙРҫ СҶРөР»РөРІРҫР№ СҮР°СҒСӮРҫСӮСӢ
        if orig_sr != Config.SR:
            audio = librosa.resample(y=audio, orig_sr=orig_sr, target_sr=Config.SR)

        audio_len = len(audio)

        # Р•СҒР»Рё С„Р°Р№Р» РҙР»РёРҪРҪСӢР№, РұРөСҖРөРј СҒР»СғСҮР°Р№РҪСӢР№ СҒСҖРөР·
        if audio_len > Config.CHUNK_SAMPLES:
            max_start = audio_len - Config.CHUNK_SAMPLES
            start = np.random.randint(0, max_start)
            audio = audio[start : start + Config.CHUNK_SAMPLES]
        elif audio_len < Config.CHUNK_SAMPLES:
            # Р•СҒР»Рё РәРҫСҖРҫСӮРәРёР№ - Р·Р°СҶРёРәР»РёРІР°РөРј СҒРҫРҙРөСҖР¶РёРјРҫРө (np.tile), Р° РҪРө РҝР°РҙРҙРёРј РҪСғР»СҸРјРё.
            # РҳРҪР°СҮРө РәРҫСҖРҫСӮРәРёР№ drone-С„Р°Р№Р» РҝСҖРөРІСҖР°СүР°Р»СҒСҸ РұСӢ РІ В«РәРҫСҖРҫСӮРәРёР№ СҒРёРіРҪР°Р» + СӮРёСҲРёРҪР°В»,
            # СҮСӮРҫ Р»РҫРјР°РөСӮ СҖР°СҒРҝСҖРөРҙРөР»РөРҪРёРө MFCC/RMS РҫСӮРҪРҫСҒРёСӮРөР»СҢРҪРҫ РҪРҫСҖРјР°Р»СҢРҪСӢС… СҒРөРәСғРҪРҙ.
            repeats = (Config.CHUNK_SAMPLES + audio_len - 1) // audio_len
            audio = np.tile(audio, repeats)[: Config.CHUNK_SAMPLES]

        # Р’РҫР·РІСҖР°СүР°РөРј СҮРёСҒСӮСӢРө numpy-РҫРұСҠРөРәСӮСӢ. РўРөРҝРөСҖСҢ РҫСҲРёРұРәРё EagerTensor РҪРө РұСғРҙРөСӮ!
        yield audio, np.float32(label)

# РӨСғРҪРәСҶРёСҸ РҙР»СҸ СҒРҫР·РҙР°РҪРёСҸ РҙР°СӮР°СҒРөСӮР° РёР· HF-РҙР°СӮР°СҒРөСӮР°
def create_dataset_from_hf(dataset, is_training=True):
    main_ds = tf.data.Dataset.from_generator(
        lambda: dataset_generator(dataset),
        output_signature=(
            tf.TensorSpec(shape=(Config.CHUNK_SAMPLES,), dtype=tf.float32),
            tf.TensorSpec(shape=(), dtype=tf.float32)
        )
    )

     # 2. РЎРҫР·РҙР°РөРј РҝРҫСӮРҫРә С„РҫРҪРҫРІРҫРіРҫ СҲСғРјР° РҙР»СҸ Р°СғРіРјРөРҪСӮР°СҶРёРё
    if is_training:
        noise_subset = dataset.filter(lambda x: x['label'] == 0)

        bg_noise_ds = tf.data.Dataset.from_generator(
            lambda: dataset_generator(noise_subset),
            output_signature=(
                tf.TensorSpec(shape=(Config.CHUNK_SAMPLES,), dtype=tf.float32),
                tf.TensorSpec(shape=(), dtype=tf.float32)
            )
        ).map(lambda audio, label: audio) # РһСӮРұСҖР°СҒСӢРІР°РөРј РјРөСӮРәСғ, РҫСҒСӮР°РІР»СҸРөРј СӮРҫР»СҢРәРҫ Р·РІСғРә

        # Р”РөР»Р°РөРј РҝРҫСӮРҫРә РұРөСҒРәРҫРҪРөСҮРҪСӢРј Рё РҝРөСҖРөРјРөСҲРёРІР°РөРј
        bg_noise_ds = bg_noise_ds.shuffle(buffer_size=1000).repeat()

        # РЎСҲРёРІР°РөРј РҫСҒРҪРҫРІРҪРҫР№ РҝРҫСӮРҫРә Рё РҝРҫСӮРҫРә СҲСғРјРҫРІ
        zipped_ds = tf.data.Dataset.zip((main_ds, bg_noise_ds))
    else:
        # Р”Р»СҸ РІР°Р»РёРҙР°СҶРёРё СҒРҫР·РҙР°РөРј РұРөСҒРәРҫРҪРөСҮРҪСӢР№ РіРөРҪРөСҖР°СӮРҫСҖ РҪСғР»РөР№ (РҝСғСҒСӮСӢСҲРәСғ)
        dummy_noise = tf.data.Dataset.from_tensors(tf.zeros(Config.CHUNK_SAMPLES, dtype=tf.float32)).repeat()
        zipped_ds = tf.data.Dataset.zip((main_ds, dummy_noise))

    # 3. РҹСҖРёРјРөРҪСҸРөРј Р°СғРіРјРөРҪСӮР°СҶРёСҺ Рё РёР·РІР»РөРәР°РөРј РҝСҖРёР·РҪР°РәРё
    # РҳСҒРҝРҫР»СҢР·СғРөРј lambda, СҮСӮРҫРұСӢ РҝРөСҖРөРҙР°СӮСҢ С„Р»Р°Рі is_training
    final_ds = zipped_ds.map(
        lambda target, noise: augment_and_extract(target, noise, is_training=is_training),
        num_parallel_calls=tf.data.AUTOTUNE
    )

    # 4. Р‘Р°СӮСҮРёРҪРі Рё РҝРөСҖРөРјРөСҲРёРІР°РҪРёРө
    if is_training:
        final_ds = final_ds.shuffle(buffer_size=5000, reshuffle_each_iteration=True)

    final_ds = final_ds.batch(Config.BATCH_SIZE).prefetch(tf.data.AUTOTUNE)

    # Р”РөР»Р°РөРј РҝРҫСӮРҫРә РұРөСҒРәРҫРҪРөСҮРҪСӢРј РҙР»СҸ СӮСҖРөРҪРёСҖРҫРІРҫСҮРҪРҫР№ РІСӢРұРҫСҖРәРё,
    # СҮСӮРҫРұСӢ РІСҖСғСҮРҪСғСҺ РәРҫРҪСӮСҖРҫР»РёСҖРҫРІР°СӮСҢ РҙР»РёРҪСғ СҚРҝРҫС…Рё СҮРөСҖРөР· steps_per_epoch
    if is_training:
        final_ds = final_ds.repeat()

    return final_ds
```

## РқРөРҝРҫСҒСҖРөРҙСҒСӮРІРөРҪРҪР°СҸ РҝРҫРҙРіРҫСӮРҫРІРәР° РҙР°РҪРҪСӢС… РҙР»СҸ РҫРұСғСҮРөРҪРёСҸ

```python
train_drones = glob.glob(f"{TRAIN_DIR}/drone/*.wav")
train_noises = glob.glob(f"{TRAIN_DIR}/noise/*.wav")
val_drones = glob.glob(f"{VAL_DIR}/drone/*.wav")
val_noises = glob.glob(f"{VAL_DIR}/noise/*.wav")

demand_path = "/kaggle/input/demand-noise-for-drone-detection/DEMAND_noise_for_drone"
# glob.glob СҒ С„Р»Р°РіРҫРј recursive=True РҪР°Р№РҙРөСӮ РІСҒРө .wav С„Р°Р№Р»СӢ РІРҫ РІСҒРөС… РҝРҫРҙРҝР°РҝРәР°С…
demand_noise_files = glob.glob(f"{demand_path}/**/*.wav", recursive=True)

train_noises.extend(demand_noise_files)

microphone_path = "/kaggle/input/datasets/shmlvssh/mircophone-dataset-for-drone-detection/MircophoneDataSet"

microphone_drone_files = glob.glob(f"{microphone_path}/drone/*.wav")
microphone_noise_files = glob.glob(f"{microphone_path}/noise/*.wav")

train_drones.extend(microphone_drone_files)
train_noises.extend(microphone_noise_files)

# РЎРёРҪСӮРөСӮРёСҮРөСҒРәР°СҸ СӮРёСҲРёРҪР° вҖ” СҸРІРҪСӢР№ РҪРөРіР°СӮРёРІРҪСӢР№ РәР»Р°СҒСҒ ----------------------------
# Р‘РөР· СҚСӮРҫРіРҫ СҒРөСӮСҢ РәР»Р°СҒСҒРёС„РёСҶРёСҖСғРөСӮ СӮРёСҲРёРҪСғ РәР°Рә В«РҙСҖРҫРҪВ»: peak-РҪРҫСҖРјР°Р»РёР·Р°СҶРёСҸ РІ РёРҪС„РөСҖРөРҪСҒРө
# СғСҒРёР»РёРІР°РөСӮ РјРёРә-С„Р»РҫСҖ РҙРҫ СғСҖРҫРІРҪСҸ СҒРёРіРҪР°Р»Р°, Р° РІ РҫРұСғСҮРөРҪРёРё РјРҫРҙРөР»СҢ СӮР°РәРҫРіРҫ СҖР°СҒРҝСҖРөРҙРөР»РөРҪРёСҸ
# РҪРө РІРёРҙРөР»Р°. Sigma РІ Р»РҫРіР°СҖРёС„РјРёСҮРөСҒРәРё СҖР°РІРҪРҫРјРөСҖРҪРҫРј РҙРёР°РҝР°Р·РҫРҪРө 1e-5..1e-2 РҝРҫРәСҖСӢРІР°РөСӮ
# РІСҒС‘ РҫСӮ РҗРҰРҹ-РјРҫР»СҮР°РҪРёСҸ РҙРҫ СӮРёС…РҫРіРҫ РҝРҫРјРөСүРөРҪРёСҸ.
print("\nрҹ”Ү Р“РөРҪРөСҖРёСҖСғРөРј СҒРёРҪСӮРөСӮРёСҮРөСҒРәСғСҺ СӮРёСҲРёРҪСғ...")
os.makedirs(Config.SILENCE_DIR, exist_ok=True)
_silence_rng = np.random.default_rng(Config.SEED)
_n_silence_samples = int(Config.SILENCE_DURATION_SEC * Config.SR)
for _i in tqdm(range(Config.NUM_SILENCE_FILES), desc="Silence"):
    _sigma = 10 ** _silence_rng.uniform(-5, -2)
    _audio = (_silence_rng.standard_normal(_n_silence_samples) * _sigma).astype(np.float32)
    sf.write(f"{Config.SILENCE_DIR}/silence_{_i:04d}.wav", _audio, Config.SR, subtype="PCM_16")

silence_files = glob.glob(f"{Config.SILENCE_DIR}/*.wav")
train_noises.extend(silence_files)
print(f"вң… Р”РҫРұР°РІР»РөРҪРҫ {len(silence_files)} С„Р°Р№Р»РҫРІ СҒРёРҪСӮРөСӮРёСҮРөСҒРәРҫР№ СӮРёСҲРёРҪСӢ РІ train_noises")

# РЎСҮРёСӮР°РөРј СҮР°РҪРәРё СҮРёСҒСӮРҫ РҙР»СҸ РёРҪС„РҫСҖРјР°СҶРёРҫРҪРҪРҫРіРҫ РІСӢРІРҫРҙР° вҖ” class_weight РұРҫР»СҢСҲРө РҪРө СҒСҮРёСӮР°РөРј,
# СӮ.Рә. balanced sampling РІ create_dataset_from_wav СҒР°Рј РҫРұРөСҒРҝРөСҮРёРІР°РөСӮ 50/50 РҪР° РұР°СӮСҮРө.
def count_total_chunks(file_paths):
    total_chunks = 0
    for path in file_paths:
        try:
            info = sf.info(path)
            # pad_end=False: С„Р°Р№Р»СӢ РәРҫСҖРҫСҮРө CHUNK_SAMPLES РҪРө РҙР°СҺСӮ РҪРё РҫРҙРҪРҫРіРҫ СҮР°РҪРәР°.
            # RMS-С„РёР»СҢСӮСҖ РҙР»СҸ drone-С„Р°Р№Р»РҫРІ РҫСӮСҖРөР¶РөСӮ РөСүС‘ РәР°РәСғСҺ-СӮРҫ РҙРҫР»СҺ вҖ” РҪРҫ СӮРҫСҮРҪРҫ
            # РҝСҖРөРҙСҒРәР°Р·Р°СӮСҢ РұРөР· СҮСӮРөРҪРёСҸ С„Р°Р№Р»Р° РҪРөР»СҢР·СҸ, РҝРҫСҚСӮРҫРјСғ СҒСҮРёСӮР°РөРј РІРөСҖС…РҪСҺСҺ РҫСҶРөРҪРәСғ.
            total_chunks += int(info.frames // Config.CHUNK_SAMPLES)
        except Exception:
            pass
    return total_chunks

print("\nвҸі РҗРҪР°Р»РёР·РёСҖСғРөРј РҝСҖРҫРҙРҫР»Р¶РёСӮРөР»СҢРҪРҫСҒСӮСҢ С„Р°Р№Р»РҫРІ...")
chunks_drone_train = count_total_chunks(train_drones)
chunks_noise_train = count_total_chunks(train_noises)
print(f"рҹ“Ҡ Р§Р°РҪРәРҫРІ РІ TRAIN (РұРөР· overlap): Р”СҖРҫРҪРҫРІ = {chunks_drone_train}, РЁСғРјР° = {chunks_noise_train}")
print(f"   Р”РёСҒРұР°Р»Р°РҪСҒ РҝРҫ СҮР°РҪРәР°Рј: {chunks_noise_train / max(chunks_drone_train, 1):.1f}x РІ РҝРҫР»СҢР·Сғ СҲСғРјР°.")
print(f"   Р‘РөР· РұР°Р»Р°РҪСҒРёСҖРҫРІРәРё РҙРҫР»СҸ РҙСҖРҫРҪРҫРІ РІ РұР°СӮСҮРө РұСӢР»Р° РұСӢ ~{100 * chunks_drone_train / (chunks_drone_train + chunks_noise_train):.1f}%.")
print(f"   sample_from_datasets РәРҫРјРҝРөРҪСҒРёСҖСғРөСӮ СҚСӮРҫ РҙРҫ 50/50 РҪР° СҮР°РҪРә-СғСҖРҫРІРҪРө.")

print("\nвҡҷпёҸ РЎРұРҫСҖРәР° РәРҫРҪРІРөР№РөСҖРҫРІ tf.data...")
train_dataset = create_dataset_from_wav(train_drones, train_noises, is_training=True)
val_dataset = create_dataset_from_wav(val_drones, val_noises, is_training=False)

print("вң… Р”Р°СӮР°СҒРөСӮСӢ РіРҫСӮРҫРІСӢ Рә РҝРҫРҙР°СҮРө РІ РҪРөР№СҖРҫСҒРөСӮСҢ!")
```

## РЎРҫР·РҙР°РҪРёРө 1D-CNN РјРҫРҙРөР»Рё

```python
# Stats pooling: concat [mean, std] РҝРҫ РҫСҒРё РІСҖРөРјРөРҪРё. Std РҙР°С‘СӮ СҒРөСӮРё feature
# СҒСӮР°СҶРёРҫРҪР°СҖРҪРҫСҒСӮРё вҖ” СҒСӮР°СҶРёРҫРҪР°СҖРҪСӢР№ РіСғР» РҙСҖРҫРҪР° РёРјРөРөСӮ РҪРёР·РәРёР№ std, РёРјРҝСғР»СҢСҒРҪСӢРө FP
# (С…Р»РҫРҝРҫРә, Р»Р°Р№) вҖ” РІСӢСҒРҫРәРёР№. GAP РІ РҫРҙРёРҪРҫСҮРәСғ СҚСӮРҫРіРҫ РҪРө СҖР°Р·Р»РёСҮР°РөСӮ.
def _stats_pool_1d(x):
    # 1. Р’СӢСҮРёСҒР»СҸРөРј СҒСҖРөРҙРҪРөРө РҝРҫ РІСҖРөРјРөРҪРё: shape (batch, channels)
    mean = GlobalAveragePooling1D()(x) 
    # 2. Р”РҫРұР°РІР»СҸРөРј РІСҖРөРјРөРҪРҪСғСҺ РҫСҒСҢ РҙР»СҸ broadcasting: shape (batch, 1, channels)
    mean_expanded = Reshape((1, -1))(mean)
    # 3. Р’СӢСҮРёСҒР»СҸРөРј (X - mean)^2
    diff = Subtract()([x, mean_expanded])
    sq_diff = Multiply()([diff, diff])
    # 4. Р”РёСҒРҝРөСҖСҒРёСҸ РәР°Рә E[(X - mean)^2]: shape (batch, channels)
    var = GlobalAveragePooling1D()(sq_diff)
    # 5. РЎСӮР°РҪРҙР°СҖСӮРҪРҫРө РҫСӮРәР»РҫРҪРөРҪРёРө СҒ Р·Р°СүРёСӮРҫР№ РҫСӮ РҪСғР»СҸ
    std = Lambda(lambda t: tf.sqrt(tf.maximum(t, 1e-6)))(var)
    
    return Concatenate(axis=-1)([mean, std])


def build_model_1d(input_shape):
    # MFCC-РІС…РҫРҙ: РҫСҒРҪРҫРІРҪРҫР№ СҒРёРіРҪР°Р» вҖ” С„РҫСҖРјР° СҒРҝРөРәСӮСҖР° РІРҫ РІСҖРөРјРөРҪРё.
    mfcc_input = Input(shape=input_shape, name="mfcc")

    # 1-Р№ РұР»РҫРә: Р‘Р°Р·РҫРІСӢРө РҝР°СӮСӮРөСҖРҪСӢ
    x = Conv1D(filters=32, kernel_size=3, padding='same')(mfcc_input)
    x = BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = MaxPooling1D(pool_size=2)(x)

    # 2-Р№ РұР»РҫРә: РЎСҖРөРҙРҪРөСғСҖРҫРІРҪРөРІСӢРө РҝСҖРёР·РҪР°РәРё
    x = Conv1D(filters=64, kernel_size=3, padding='same')(x)
    x = BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = MaxPooling1D(pool_size=2)(x)

    # 3-Р№ РұР»РҫРә: Р“Р»СғРұРҫРәРёРө Р°РәСғСҒСӮРёСҮРөСҒРәРёРө СҒСӮСҖСғРәСӮСғСҖСӢ
    x = Conv1D(filters=128, kernel_size=3, padding='same')(x)
    x = BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Stats pooling: [mean, std] РҝРҫ РІСҖРөРјРөРҪРё.
    x = _stats_pool_1d(x)

    # RMS-РІС…РҫРҙ: РІСҒРҝРҫРјРҫРіР°СӮРөР»СҢРҪСӢР№ СҒРәР°Р»СҸСҖ (log-RMS СҮР°РҪРәР°) вҖ” РҙР°С‘СӮ СҒРөСӮРё РёРҪС„РҫСҖРјР°СҶРёСҺ
    # Рҫ СҒРҫРҫСӮРҪРҫСҲРөРҪРёРё РҝРёРә/RMS (С„РҫСҖРјР° СҒРёРіРҪР°Р»Р°) РҝРҫСҒР»Рө peak-normalize.
    rms_input = Input(shape=(1,), name="rms")
    rms_features = Dense(8, activation='relu', name="rms_features")(rms_input)

    # РЎР»РёСҸРҪРёРө СҒРҝРөРәСӮСҖР°Р»СҢРҪРҫР№ Рё СғСҖРҫРІРҪРөРІРҫР№ РІРөСӮРІРөР№.
    combined = tf.keras.layers.Concatenate(axis=-1)([x, rms_features])
    combined = Dropout(0.3)(combined)

    outputs = Dense(1, activation='sigmoid', name="output_prob")(combined)

    model = Model(inputs=[mfcc_input, rms_input], outputs=outputs, name="1D_CNN")
    return model

# 1. Р’СӢСҮРёСҒР»СҸРөРј СҖР°Р·РјРөСҖРҪРҫСҒСӮСҢ РІС…РҫРҙР°
# РқР° 1 СҒРөРәСғРҪРҙСғ РҝСҖРё SR=32000, РҫРәРҪРө 1024 Рё СҲР°РіРө 512 РҝРҫР»СғСҮР°РөСӮСҒСҸ 61 С„СҖРөР№Рј.
mfcc_frames = (Config.CHUNK_SAMPLES - Config.FRAME_LENGTH) // Config.FRAME_STEP + 1
input_shape_1d = (mfcc_frames, Config.N_MFCC)

print(input_shape_1d)

# 2. РЎРҫР·РҙР°РөРј Рё РәРҫРјРҝРёР»РёСҖСғРөРј РјРҫРҙРөР»СҢ
model_1d = build_model_1d(input_shape_1d)

model_1d.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss='binary_crossentropy',
    metrics=['accuracy', Precision(name='precision'), Recall(name='recall')]
)

model_1d.summary()
```

## РһРұСғСҮРөРҪРёРө 1D-CNN РјРҫРҙРөР»Рё

```python
# 3. РқР°СҒСӮСҖРҫР№РәР° РәРҫР»Р»РұРөРәРҫРІ (Р°РІСӮРҫРјР°СӮРёСҮРөСҒРәРҫРө СҒРҫС…СҖР°РҪРөРҪРёРө Р»СғСҮСҲРөР№ РјРҫРҙРөР»Рё Рё РҫСҒСӮР°РҪРҫРІРәР°)
callbacks_1d =[
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
    ModelCheckpoint('1d_cnn.keras', monitor='val_loss', save_best_only=True, verbose=0)
]

train_dataset_1d = train_dataset.map(
    lambda features, label: ({"mfcc": features["mfcc"], "rms": features["rms"]}, label),
    num_parallel_calls=tf.data.AUTOTUNE
)
val_dataset_1d = val_dataset.map(
    lambda features, label: ({"mfcc": features["mfcc"], "rms": features["rms"]}, label),
    num_parallel_calls=tf.data.AUTOTUNE
)

history_1d = model_1d.fit(
    train_dataset_1d,
    validation_data=val_dataset_1d,
    epochs=Config.EPOCHS,
    callbacks=callbacks_1d,
    steps_per_epoch=Config.STEPS_PER_EPOCH,
)
```

## РЎРҫР·РҙР°РҪРёРө 2D-CNN РјРҫРҙРөР»Рё

```python
# РҗРҪР°Р»РҫРі _stats_pool_1d, РҪРҫ РҙР»СҸ 2D-РәР°СҖСӮ РҝСҖРёР·РҪР°РәРҫРІ.
def _stats_pool_2d(x):
    # 1. РЎСҖРөРҙРҪРөРө РҝРҫ РҝСҖРҫСҒСӮСҖР°РҪСҒСӮРІСғ: shape (batch, channels)
    mean = GlobalAveragePooling2D()(x)
    # 2. Р”РҫРұР°РІР»СҸРөРј РҝСҖРҫСҒСӮСҖР°РҪСҒСӮРІРөРҪРҪСӢРө РҫСҒРё: shape (batch, 1, 1, channels)
    mean_expanded = Reshape((1, 1, -1))(mean)
    # 3. Р’СӢСҮРёСҒР»СҸРөРј (X - mean)^2
    diff = Subtract()([x, mean_expanded])
    sq_diff = Multiply()([diff, diff])
    # 4. Р”РёСҒРҝРөСҖСҒРёСҸ
    var = GlobalAveragePooling2D()(sq_diff)
    # 5. РЎСӮР°РҪРҙР°СҖСӮРҪРҫРө РҫСӮРәР»РҫРҪРөРҪРёРө
    std = Lambda(lambda t: tf.sqrt(tf.maximum(t, 1e-6)))(var)
    
    return Concatenate(axis=-1)([mean, std])

def build_model_2d(input_shape):
    inputs = Input(shape=input_shape, name="mel_spectrogram")

    # 1. РҹРөСҖРІСӢР№ СҒР»РҫР№ - РҫРұСӢСҮРҪР°СҸ СҒРІРөСҖСӮРәР°
    x = Conv2D(32, kernel_size=(3, 3), padding='same', use_bias=False)(inputs)
    x = BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)

    # 2. Р‘Р»РҫРә 1 (РЎРөРҝР°СҖР°РұРөР»СҢРҪСӢР№)
    x = SeparableConv2D(64, kernel_size=(3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)

    # 3. Р‘Р»РҫРә 2 (РЎРөРҝР°СҖР°РұРөР»СҢРҪСӢР№)
    x = SeparableConv2D(128, kernel_size=(3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = MaxPooling2D(pool_size=(2, 2))(x)

    # 4. Р‘Р»РҫРә 3 (РіР»СғРұРҫРәРёР№)
    x = SeparableConv2D(256, kernel_size=(3, 3), padding='same', use_bias=False)(x)
    x = BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)

    # Stats pooling: [mean, std] РҝРҫ РҝСҖРҫСҒСӮСҖР°РҪСҒСӮРІРөРҪРҪСӢРј РҫСҒСҸРј.
    x = _stats_pool_2d(x)

    x = Dropout(0.3)(x)

    outputs = Dense(1, activation='sigmoid', name="output_prob")(x)

    model = Model(inputs=inputs, outputs=outputs, name="2D_CNN")
    return model


# 1. Р’СӢСҮРёСҒР»СҸРөРј СҖР°Р·РјРөСҖРҪРҫСҒСӮСҢ РІС…РҫРҙР°
# Р”РҫРұР°РІР»СҸРөРј "1" РІ РәРҫРҪСҶРө, СӮР°Рә РәР°Рә РңРөР»-СҒРҝРөРәСӮСҖРҫРіСҖР°РјРјР° - СҚСӮРҫ РҫРҙРҪРҫРәР°РҪР°Р»СҢРҪРҫРө РёР·РҫРұСҖР°Р¶РөРҪРёРө (Grayscale)
input_shape_2d = (mfcc_frames, Config.N_MELS, 1)

print(input_shape_2d)

# 2. РЎРҫР·РҙР°РөРј Рё РәРҫРјРҝРёР»РёСҖСғРөРј РјРҫРҙРөР»СҢ
model_2d = build_model_2d(input_shape_2d)

# Р”Р»СҸ РІСӮРҫСҖРҫР№ РјРҫРҙРөР»Рё РјСӢ РјРҫР¶РөРј РёСҒРҝРҫР»СҢР·РҫРІР°СӮСҢ СҮСғСӮСҢ РјРөРҪСҢСҲРёР№ learning rate
model_2d.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.0005),
    loss='binary_crossentropy',
    metrics=['accuracy', Precision(name='precision'), Recall(name='recall')]
)

model_2d.summary()
```

## РһРұСғСҮРөРҪРёРө 2D-CNN РјРҫРҙРөР»Рё

```python
# 3. РқР°СҒСӮСҖРҫР№РәР° РәРҫР»Р»РұРөРәРҫРІ
callbacks_2d =[
    EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True, verbose=1),
    ModelCheckpoint('2d_cnn.keras', monitor='val_loss', save_best_only=True, verbose=0)
]

train_dataset_2d = train_dataset.map(
    lambda features, label: (features["mel_spectrogram"], label),
    num_parallel_calls=tf.data.AUTOTUNE
)
val_dataset_2d = val_dataset.map(
    lambda features, label: (features["mel_spectrogram"], label),
    num_parallel_calls=tf.data.AUTOTUNE
)

history_2d = model_2d.fit(
    train_dataset_2d,
    validation_data=val_dataset_2d,
    epochs=Config.EPOCHS,
    callbacks=callbacks_2d,
    steps_per_epoch=Config.STEPS_PER_EPOCH,
)
```

## Р”РҫРұР°РІР»РөРҪРёРө СҒР»РҫСҸ СҒ РҝСҖРөРҙРҫРұСҖР°РұРҫСӮРәРҫР№ РҙР°РҪРҪСӢС… РІ СҒР°РјРё РјРҫРҙРөР»Рё

```python
def representative_dataset_1d():
    # 1D-CNN РҙРІСғС…РІС…РҫРҙРҫРІР°СҸ (mfcc + rms). РҹРөСҖРөРҙР°С‘Рј СҒРҝРёСҒРҫРә РІ РҝРҫСҖСҸРҙРәРө model.inputs;
    # dict Р·РҙРөСҒСҢ РјРҫР»СҮР° Р»РҫРјР°РөСӮ РәР°Р»РёРұСҖРҫРІРәСғ INT8.
    for features, _ in train_dataset.unbatch().take(500):
        yield [
            tf.expand_dims(features["mfcc"], 0),
            tf.expand_dims(features["rms"], 0),
        ]

def representative_dataset_2d():
    for features, _ in train_dataset.unbatch().take(500):
        # Р”РҫРұР°РІР»СҸРөРј СҖР°Р·РјРөСҖРҪРҫСҒСӮСҢ РұР°СӮСҮР° -> (1, NUM_FRAMES, N_MELS, 1)
        yield [tf.expand_dims(features["mel_spectrogram"], 0)]

def _print_tflite_io(model_path):
    """Р’СӢРІРҫРҙРёСӮ СҒС…РөРјСғ РІС…РҫРҙРҫРІ/РІСӢС…РҫРҙРҫРІ TFLite-РјРҫРҙРөР»Рё РҙР»СҸ РІРөСҖРёС„РёРәР°СҶРёРё.

    РҡСҖРёСӮРёСҮРҪРҫ РҙР»СҸ 1D-РјРҫРҙРөР»Рё: app/main.py РёСүРөСӮ РёРҪРҙРөРәСҒСӢ РІС…РҫРҙРҫРІ РҝРҫ РҝРҫРҙСҒСӮСҖРҫРәР°Рј
    'mfcc' Рё 'rms' РІ РёРјРөРҪРё СӮРөРҪР·РҫСҖР°. Р•СҒР»Рё РәРҫРҪРІРөСҖСӮРөСҖ РҝРҫСӮРөСҖСҸР» РёРјРөРҪР° РёР»Рё
    РјРҫРҙРөР»СҢ РҫРәР°Р·Р°Р»Р°СҒСҢ РҫРҙРҪРҫРІС…РҫРҙРҫРІРҫР№ вҖ” main.py СғРҝР°РҙС‘СӮ РҪР° StopIteration
    РІ next(d for d in details if "mfcc" in d["name"]).
    """
    interp = tf.lite.Interpreter(model_path=model_path)
    interp.allocate_tensors()
    in_details = interp.get_input_details()
    out_details = interp.get_output_details()

    print(f"   в”Җв”Җ TFLite I/O РҙР»СҸ {model_path} в”Җв”Җ")
    print(f"   Р’С…РҫРҙРҫРІ: {len(in_details)}")
    for d in in_details:
        print(
            f"     [{d['index']}] name='{d['name']}' "
            f"shape={tuple(d['shape'])} dtype={np.dtype(d['dtype']).name}"
        )
    print(f"   Р’СӢС…РҫРҙРҫРІ: {len(out_details)}")
    for d in out_details:
        print(
            f"     [{d['index']}] name='{d['name']}' "
            f"shape={tuple(d['shape'])} dtype={np.dtype(d['dtype']).name}"
        )

    # Sanity-check РҙР»СҸ 1D-РјРҫРҙРөР»Рё: СҖРҫРІРҪРҫ 2 РІС…РҫРҙР°, РёРјРөРҪР° СҒРҫРҙРөСҖР¶Р°СӮ 'mfcc' Рё 'rms'.
    if "1d" in model_path.lower():
        names = [d["name"].lower() for d in in_details]
        has_mfcc = any("mfcc" in n for n in names)
        has_rms = any("rms" in n for n in names)
        if len(in_details) == 2 and has_mfcc and has_rms:
            print("   вң… 1D-РјРҫРҙРөР»СҢ: 2 РІС…РҫРҙР° (mfcc + rms) РҫРҝСҖРөРҙРөР»РөРҪСӢ РәРҫСҖСҖРөРәСӮРҪРҫ.")
        else:
            print(
                f"   вқҢ Р’РқРҳРңРҗРқРҳР•: РҫР¶РёРҙР°Р»РёСҒСҢ 2 РІС…РҫРҙР° СҒ РёРјРөРҪР°РјРё 'mfcc' Рё 'rms', "
                f"РҝРҫР»СғСҮРөРҪРҫ {len(in_details)}: {names}. "
                f"app/main.py РҪРө СҒРјРҫР¶РөСӮ РҪР°Р№СӮРё СӮРөРҪР·РҫСҖСӢ РҝРҫ РҝРҫРҙСҒСӮСҖРҫРәР°Рј!"
            )


def convert_pure_cnn_int8(concrete_fn, rep_gen, output_filename):
    print(f"вҸі РҡРҫРҪРІРөСҖСӮР°СҶРёСҸ {output_filename}...")
    converter = tf.lite.TFLiteConverter.from_concrete_functions([concrete_fn])

    # РўРҫР»СҢРәРҫ СҒСӮР°РҪРҙР°СҖСӮРҪСӢРө РҫРҝРөСҖР°СҶРёРё TFLite (СҖР°РұРҫСӮР°РөСӮ Р’Р•Р—Р”Р•!)
    converter.target_spec.supported_ops = [
        tf.lite.OpsSet.TFLITE_BUILTINS_INT8,
        tf.lite.OpsSet.TFLITE_BUILTINS
    ]

    # РҡРІР°РҪСӮРҫРІР°РҪРёРө
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    converter.representative_dataset = rep_gen

    # РҹРҫР»РҪРҫРө INT8 РәРІР°РҪСӮРҫРІР°РҪРёРө
    converter.target_spec.supported_types = [tf.int8]
    converter.inference_input_type = tf.float32
    converter.inference_output_type = tf.float32

    tflite_model = converter.convert()

    with open(output_filename, 'wb') as f:
        f.write(tflite_model)
    print(f"вң… РЎРҫС…СҖР°РҪРөРҪРҫ: {output_filename} (Р Р°Р·РјРөСҖ: {os.path.getsize(output_filename)/1024:.2f} KB)")

    # Р’РөСҖРёС„РёРәР°СҶРёСҸ: СҖР°СҒРҝРөСҮР°СӮР°СӮСҢ СҒС…РөРјСғ РІС…РҫРҙРҫРІ/РІСӢС…РҫРҙРҫРІ РіРҫСӮРҫРІРҫРіРҫ .tflite.
    _print_tflite_io(output_filename)


# Concrete function СҒ РҙРІСғРјСҸ РёРјРөРҪРҫРІР°РҪРҪСӢРјРё РҝРҫР·РёСҶРёРҫРҪРҪСӢРјРё РІС…РҫРҙР°РјРё РҙР»СҸ 1D-CNN.
# РҳРјРөРҪР° 'mfcc' Рё 'rms' СҒРҫС…СҖР°РҪСҸСҺСӮСҒСҸ РІ TFLite-СҒРёРіРҪР°СӮСғСҖРө, РҝРҫ РҪРёРј main.py РёСүРөСӮ
# РёРҪРҙРөРәСҒСӢ СӮРөРҪР·РҫСҖРҫРІ (next(d for d in details if "mfcc" in d["name"])).
@tf.function(input_signature=[
    tf.TensorSpec(shape=[1, Config.NUM_FRAMES, Config.N_MFCC], dtype=tf.float32, name='mfcc'),
    tf.TensorSpec(shape=[1, 1], dtype=tf.float32, name='rms'),
])
def serve_1d(mfcc, rms):
    return model_1d([mfcc, rms])

# Р”Р»СҸ 2D РІС…РҫРҙ РҫРҙРёРҪ, РҪРҫ РёСҒРҝРҫР»СҢР·СғРөРј СӮРҫСӮ Р¶Рө РҝСғСӮСҢ СҮРөСҖРөР· concrete function СҖР°РҙРё
# РөРҙРёРҪРҫРҫРұСҖР°Р·РёСҸ Рё СҮСӮРҫРұСӢ Р·Р°С„РёРәСҒРёСҖРҫРІР°СӮСҢ batch=1 РІ СҒРёРіРҪР°СӮСғСҖРө.
@tf.function(input_signature=[
    tf.TensorSpec(shape=[1, Config.NUM_FRAMES, Config.N_MELS, 1], dtype=tf.float32, name='mel_spectrogram'),
])
def serve_2d(mel_spectrogram):
    return model_2d(mel_spectrogram)

# 3. Р—Р°РҝСғСҒРә
convert_pure_cnn_int8(serve_1d.get_concrete_function(), representative_dataset_1d, "model_1d.tflite")
convert_pure_cnn_int8(serve_2d.get_concrete_function(), representative_dataset_2d, "model_2d.tflite")
```

```python
# ==========================================
# Р‘РӣРһРҡ 6: РһРҰР•РқРҡРҗ Р Р•Р—РЈРӣР¬РўРҗРўРһР’ Рҳ Р’РҳР—РЈРҗРӣРҳР—РҗРҰРҳРҜ
# ==========================================

# ---------------------------------------------------------
# 6.1: РӨСғРҪРәСҶРёСҸ РҙР»СҸ РҝРҫСҒСӮСҖРҫРөРҪРёСҸ РіСҖР°С„РёРәРҫРІ РҫРұСғСҮРөРҪРёСҸ
# ---------------------------------------------------------
def plot_learning_curves(history, model_name):
    """
    РҹСҖРёРҪРёРјР°РөСӮ РҫРұСҠРөРәСӮ history РҫСӮ model.fit() Рё СҒСӮСҖРҫРёСӮ РіСҖР°С„РёРәРё.
    """
    # РЎРҫР·РҙР°РөРј РҫРұР»Р°СҒСӮСҢ РҙР»СҸ РҙРІСғС… РіСҖР°С„РёРәРҫРІ СҖСҸРҙРҫРј
    plt.figure(figsize=(16, 6))

    # --- Р“СҖР°С„РёРә 1: РҹРҫСӮРөСҖРё (Loss) ---
    plt.subplot(1, 2, 1)
    plt.plot(history.history['loss'], label='РҹРҫСӮРөСҖРё РҪР° РҫРұСғСҮРөРҪРёРё (Train Loss)')
    plt.plot(history.history['val_loss'], label='РҹРҫСӮРөСҖРё РҪР° РІР°Р»РёРҙР°СҶРёРё (Val Loss)')
    plt.title(f'Р“СҖР°С„РёРә РҝРҫСӮРөСҖСҢ РҙР»СҸ РјРҫРҙРөР»Рё "{model_name}"')
    plt.xlabel('РӯРҝРҫС…Р°')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    # --- Р“СҖР°С„РёРә 2: РңРөСӮСҖРёРәРё (Accuracy, Precision, Recall) ---
    plt.subplot(1, 2, 2)
    plt.plot(history.history['accuracy'], label='РўРҫСҮРҪРҫСҒСӮСҢ (Train Accuracy)')
    plt.plot(history.history['val_accuracy'], label='РўРҫСҮРҪРҫСҒСӮСҢ (Val Accuracy)', linestyle='--')
    plt.plot(history.history['precision'], label='Precision (Train)')
    plt.plot(history.history['val_precision'], label='Precision (Val)', linestyle='--')
    plt.plot(history.history['recall'], label='Recall (Train)')
    plt.plot(history.history['val_recall'], label='Recall (Val)', linestyle='--')
    plt.title(f'Р“СҖР°С„РёРә РјРөСӮСҖРёРә РҙР»СҸ РјРҫРҙРөР»Рё "{model_name}"')
    plt.xlabel('РӯРҝРҫС…Р°')
    plt.ylabel('Р—РҪР°СҮРөРҪРёРө РјРөСӮСҖРёРәРё')
    plt.legend()
    plt.grid(True)

    plt.tight_layout() # Р§СӮРҫРұСӢ РіСҖР°С„РёРәРё РҪРө РҪР°РөР·Р¶Р°Р»Рё РҙСҖСғРі РҪР° РҙСҖСғРіР°
    plt.show()

# ---------------------------------------------------------
# 6.2: Р’СӢР·РҫРІ С„СғРҪРәСҶРёР№ РҙР»СҸ РҫРұРөРёС… РјРҫРҙРөР»РөР№
# ---------------------------------------------------------
# Р’РқРҳРңРҗРқРҳР•: Р•СҒР»Рё СӮСӢ РҝРөСҖРөР·Р°РҝСғСҒСӮРёР» РҪРҫСғСӮРұСғРә, СғРұРөРҙРёСҒСҢ, СҮСӮРҫ РҝРөСҖРөРјРөРҪРҪСӢРө
# history_1d Рё history_2d СҒСғСүРөСҒСӮРІСғСҺСӮ РҝРҫСҒР»Рө РҫРұСғСҮРөРҪРёСҸ!
if 'history_1d' in locals():
    print("рҹ“Ҳ Р’РёР·СғР°Р»РёР·Р°СҶРёСҸ СҖРөР·СғР»СҢСӮР°СӮРҫРІ РҙР»СҸ 1D-CNN")
    plot_learning_curves(history_1d, "1D-CNN")
else:
    print("вҡ пёҸ РһРұСҠРөРәСӮ history_1d РҪРө РҪР°Р№РҙРөРҪ. РҹСҖРҫРҝСғСҒСӮРёР»Рё РҫРұСғСҮРөРҪРёРө РңРҫРҙРөР»Рё 1.")

if 'history_2d' in locals():
    print("\nрҹ“Ҳ Р’РёР·СғР°Р»РёР·Р°СҶРёСҸ СҖРөР·СғР»СҢСӮР°СӮРҫРІ РҙР»СҸ 2D-CNN")
    plot_learning_curves(history_2d, "2D-CNN")
else:
    print("вҡ пёҸ РһРұСҠРөРәСӮ history_2d РҪРө РҪР°Р№РҙРөРҪ. РҹСҖРҫРҝСғСҒСӮРёР»Рё РҫРұСғСҮРөРҪРёРө РңРҫРҙРөР»Рё 2.")

# ---------------------------------------------------------
# 6.3: Р’РёР·СғР°Р»РёР·Р°СҶРёСҸ Р°СҖС…РёСӮРөРәСӮСғСҖСӢ РјРҫРҙРөР»РөР№ (Р“СҖР°С„СӢ)
# ---------------------------------------------------------
# РӯСӮР° С„СғРҪРәСҶРёСҸ СҒРҫС…СҖР°РҪРёСӮ СҒС…РөРјСӢ РјРҫРҙРөР»РөР№ РІ РІРёРҙРө PNG С„Р°Р№Р»РҫРІ РҝСҖСҸРјРҫ РІ Kaggle.
# РӯСӮРё РәР°СҖСӮРёРҪРәРё СӮСӢ СҒРјРҫР¶РөСҲСҢ СҒРәР°СҮР°СӮСҢ Рё РІСҒСӮР°РІРёСӮСҢ РІ СҒРІРҫСҺ РҙРёСҒСҒРөСҖСӮР°СҶРёСҺ.

try:
    # Р“СҖР°С„ РҙР»СҸ 1D-CNN
    plot_model(
        model_1d,
        to_file='architecture_1d_cnn.png',
        show_shapes=True, # РһРұСҸР·Р°СӮРөР»СҢРҪРҫ РҝРҫРәР°Р·СӢРІР°РөРј СҖР°Р·РјРөСҖРҪРҫСҒСӮРё СӮРөРҪР·РҫСҖРҫРІ!
        show_layer_names=True,
        rankdir='TB' # Р’РөСҖСӮРёРәР°Р»СҢРҪР°СҸ РҫСҖРёРөРҪСӮР°СҶРёСҸ (Top-to-Bottom)
    )
    print("\nвң… РЎС…РөРјР° Р°СҖС…РёСӮРөРәСӮСғСҖСӢ РңРҫРҙРөР»Рё 1 СҒРҫС…СҖР°РҪРөРҪР° РІ 'architecture_1d_cnn.png'")

    # Р“СҖР°С„ РҙР»СҸ 2D-CNN
    plot_model(
        model_2d,
        to_file='architecture_2d_cnn.png',
        show_shapes=True,
        show_layer_names=True,
        rankdir='TB'
    )
    print("вң… РЎС…РөРјР° Р°СҖС…РёСӮРөРәСӮСғСҖСӢ РңРҫРҙРөР»Рё 2 СҒРҫС…СҖР°РҪРөРҪР° РІ 'architecture_2d_cnn.png'")

except Exception as e:
    print(f"\nвқҢ РқРө СғРҙР°Р»РҫСҒСҢ СҒРіРөРҪРөСҖРёСҖРҫРІР°СӮСҢ СҒС…РөРјСӢ. Р’РҫР·РјРҫР¶РҪРҫ, РҪСғР¶РҪРҫ СғСҒСӮР°РҪРҫРІРёСӮСҢ 'pydot' Рё 'graphviz'.")
    print(f"РҹРҫРҝСҖРҫРұСғР№СӮРө РІСӢРҝРҫР»РҪРёСӮСҢ РІ СҸСҮРөР№РәРө: !pip install pydot graphviz")
```
